```python
from data_utils import TextMelLoader
import pandas as pd
import hparams

hparams = hparams.create_hparams()

speakers_trn = pd.read_csv('filelists/libritts_speakerinfo.txt', 
                           engine='python',header=None, comment=';', sep=' *\| *', 
                           names=['ID', 'SEX', 'SUBSET', 'MINUTES', 'NAME'])
speakers_evl = pd.read_csv('filelists/libritts_speakerinfo.txt', 
                           engine='python',header=None, comment=';', sep=' *\| *', 
                           names=['ID', 'SEX', 'SUBSET', 'MINUTES', 'NAME'])

# calculate mellotron ids using evaluation filelist
speaker_ids_evl = TextMelLoader("filelists/libritts_train_clean_100_audiopath_text_sid_atleast5min_val_filelist.txt", 
                            hparams).speaker_ids
speakers_evl['MELLOTRON_ID'] = speakers_evl['ID'].apply(lambda x: speaker_ids_evl[x] if x in speaker_ids_evl else -1)

# calculate mellotron ids using training filelist
speaker_ids_trn = TextMelLoader("filelists/libritts_train_clean_100_audiopath_text_sid_shorterthan10s_atleast5min_train_filelist.txt",
                            hparams).speaker_ids
speakers_trn['MELLOTRON_ID'] = speakers_trn['ID'].apply(lambda x: speaker_ids_trn[x] if x in speaker_ids_trn else -1)

print("# of eval ids:{}\n# of train ids:{}\n\n".format(
    len(speakers_evl.query("MELLOTRON_ID >= 0")),
    len(speakers_trn.query("MELLOTRON_ID >= 0"))))

print("{:13} {:15} {:25} {}".format("LibriTTS_Id", "Mellotron_Id", "Speaker Name", "Speaker Name"))
print("{:5} {:7} {:5} {:9} {:25} {}".format(
    "Eval", "Train", "Eval", "Train", "Eval", "Train"))
print("--------------------------------------------------------------------")
for x, y in zip(speakers_evl.query("MELLOTRON_ID >= 0").iterrows(),
                speakers_trn.query("MELLOTRON_ID >= 0").iterrows()):
    print("{:<5} {:<7} {:<5} {:<9} {:25} {}".format(
        x[1]['ID'], y[1]['ID'], x[1]['MELLOTRON_ID'], y[1]['MELLOTRON_ID'], x[1]['NAME'], y[1]['NAME']))
          
```

```python
speaker_ids = TextMelLoader("filelists/libritts_train_clean_100_audiopath_text_sid_shorterthan10s_atleast5min_train_filelist.txt",
                            hparams).speaker_ids
speakers_trn['MELLOTRON_ID'] = speakers_trn['ID'].apply(lambda x: speaker_ids[x] if x in speaker_ids else -1)
```
